```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
import numpy as np

```

```python
df =pd.read_csv('titanic.csv')
```

```python
df = df.drop(columns = ['boat', 'body', 'cabin', 'home.dest'])
```

```python
df = df.dropna()
```

one hot encoding

```python
df = pd.get_dummies(df, columns = ['sex', 'embarked'])
```

reproducibility

```python
random = 42
np.random.seed(42)
```

cross validation

```python
train,test = train_test_split(df, test_size = 0.1, random_state = random)
train, val = train_test_split(train, test_size = 0.2, random_state = random)
```

```python
def evaluate(test,predictions):
  true_pos = 0
  true_neg = 0
  false_pos = 0
  false_neg = 0
  for i, j in zip(list(test['survived']),predictions):
    if i ==1:
      if j == 1:
        true_pos += 1
      else:
        false_neg += 1
    else:
      if j == 1:
        false_pos += 1
      else:
        true_neg += 1

  # Indentation corrected: Calculate and return metrics outside the loop
  precision = true_pos / (true_pos + false_pos) if (true_pos + false_pos) != 0 else 0  # Handle ZeroDivisionError
  recall = true_pos / (true_pos + false_neg) if (true_pos + false_neg) != 0 else 0  # Handle ZeroDivisionError
  f1 = 2 * precision * recall / (precision + recall) if (precision + recall) != 0 else 0  # Handle ZeroDivisionError
  accuracy = (true_pos + true_neg) / len(list(test['survived']))


  return precision, recall, f1, accuracy
```

```python
predictions = [1] * len(val)
precision, recall, f1, accuracy = evaluate(val, predictions)
print(precision)
print(recall)
print(f1)
print(accuracy)
```

```python
accuracy_score(list(val['survived']), predictions)
```

```python
f1_score(list(val['survived']),predictions)
```
